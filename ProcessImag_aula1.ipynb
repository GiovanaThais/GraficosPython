```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pylab as plt
```

```python
#carregando imagem
#imageurl = "https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Cat03.jpg/600px-Cat03.jpg?20160327035619"
imageurl = "https://upload.wikimedia.org/wikipedia/commons/thumb/6/65/Blue_morpho_butterfly.jpg/663px-Blue_morpho_butterfly.jpg"
#imageurl = "https://c.wallhere.com/photos/3d/2b/parrots_birds_branch_sit_colorful-645199.jpg"

image = io.imread(imageurl)

cv2_imshow(image)
```

```python
print(image.shape)
```

```python
(h, w, c) = image.shape
print(f'Largura: {w} pixels')
print(f'Altura: {h} pixels')
print(f'Canais: {c} pixels')
```

```python
from skimage.color import rgb2gray, rgb2gray

#converter a imagem para niveis de cinza
img_gray = rgb2gray(image)
plt.imshow(img_gray, cmap = 'gray')
plt.show()
```

Redimencionar e rotacionar uma imagem 

```python
from skimage.transform import resize, rotate

img_r = image[:, :,0]
plt.imshow(img_r)
plt.show()

img_g = image[:, :,2]
plt.imshow(img_r)
plt.show()

print('Maximo de pixels: ')
print(img_g.max())

print('Minimo de pixels: ')
print(img_g.min())

print('Média de pixels: ')
print(img_g.mean())

```

```python
a, l, c = image.shape
img_red2 = resize(image, (int(a*2), int(l*2)))
plt.imshow(img_red2)
plt.show()
```

```python
a, l, c = image.shape
img_red1 = resize(image, (int(a/2), int(l/2)))
plt.imshow(img_red1)
plt.show()
```

```python
img_rot = rotate(image, 180, resize=True)
plt.imshow(img_rot)
plt.show()
```

Exibir todos os pixels da imagem em um histograma

```python
plt.hist(image.ravel(), bins=256, range=[0,256])
plt.show()
```

12 Etapa

transformando em tons de cinza(operação inversa)

```python
new_image1 = 255 - gray_image
cv2_imshow(new_image1)
```

```python
new_image2 = (100.0/255)*gray_image +100
cv2_imshow(new_image2)
```

```python
new_image3 = 255.0*(gray_image/255.0)**2
cv2_imshow(new_image3)
```
